```python
def contains_num2(x):
  values=[]
  
  string = [*x]

  for i in range(len(string)):
    if unicodedata.digit(string[i],-1) == -1:
        values.append(False)
    elif unicodedata.digit(string[i],-1) != -1:
        values.append(True)

  if True in values:
    return 1
  else:
    return 0
```
